## Mathematics+Plot+Library = Matplotlib

```python
# First Code In Matplotlib

# importing the module and remane it as plt.
import matplotlib.pyplot as plt

# plot function is used to plot the graph.
# the first list is the x-axis and the second list is the y-axis.
# the first point is (1, 1), the second point is (2, 4), the third point is (3, 9) and the fourth point is (4, 6).
# Syntax: plt.plot(x-axis, y-axis)
plt.plot([1, 2, 3, 4], 
         [1, 4, 9, 6])
# show function is used to show the graph.
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

ypoints = np.array([3, 8, 1, 10])

plt.figure()
# plt.plot(ypoints, 'r') // r = red
plt.plot(ypoints, '-.g') # g = green, -. = dash-dot
plt.show()
```
