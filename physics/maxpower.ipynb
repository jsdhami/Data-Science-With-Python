```python
# creating line graph
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# import data 
data = pd.read_csv('../data/maxpower.csv')
# print just top datas
print(data.head())
```

```python
# plot graph between R and P1

plt.plot(data['R'], data['P1'], label='P1')
# draw also horizontal line
max_power1 = data['P1'].max()
plt.axhline(y=max_power1, color='g', linestyle='--')

# draw max line in graph, take value when P2 is max
max_power2 = data['R'][data['P1'].idxmax()]
plt.axvline(x=max_power2, color='r', linestyle='--')
# labelling
plt.xlabel('Resistance-R')
plt.ylabel('Power By Using I and R')
plt.title('Maximum Power Theorem')
plt.legend()

plt.show()

# Maximum power from P1
max_power1 = data['P1'].max()
print('Maximum power from V and R:', max_power1)
```

```python
# plot graph between R and P2

plt.plot(data['R'], data['P2'], label='P2')

# draw also horizontal line
max_power1 = data['P2'].max()
plt.axhline(y=max_power1, color='g', linestyle='--')

# draw max line in graph, take value when P2 is max
max_power2 = data['R'][data['P2'].idxmax()]
plt.axvline(x=max_power2, color='r', linestyle='--')

# labelling
plt.xlabel('Resistance-R')
plt.ylabel('Power By Using V and R')
plt.title('Maximum Power Theorem')
plt.legend()

plt.show()

# maximum power from P2
max_power2 = data['P2'].max()
print('Maximum power from I and R:', max_power2)
```
